```python
train_dataframe=pd.read_csv("train.csv")
test_dataframe=pd.read_csv("test.csv")
train_dataframe
```

```python
unique_nationalities = test_dataframe['Nacionality'].value_counts()
unique_nationalities
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
# Assuming df is your DataFrame
df = train_dataframe

# Define features and target variable, excluding the 'id' column
X = df.drop(columns=['Target', 'id'])
y = df['Target']
```

```python
# Initialize and train the Random Forest classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X, y)
```

```python
# Make predictions on the test set
y_pred_encoded = clf.predict(X)
```

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y, y_pred_encoded))
print("\nClassification Report:")
print(classification_report(y, y_pred_encoded))
print("\nAccuracy Score:")
print(accuracy_score(y, y_pred_encoded))
```

```python
# Define features for testing (assuming test dataframe doesn't have target column)
X_test = test_dataframe

X_test = X_test.drop(columns=['id'])
```

```python
# Make predictions on the test set
y_pred = clf.predict(X_test)
```

```python
# Add the predicted target column to the test dataframe
test_dataframe['Target'] = y_pred
test_dataframe
```

```python
# Select only the 'id' and 'Target' columns
filtered_df = test_dataframe[['id', 'Target']]
filtered_df
```

```python
# Save the test dataframe with predicted target to a CSV file
filtered_df.to_csv('submit.csv', index=False)
```
